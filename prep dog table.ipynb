```python
import pandas as pd
dogs = pd.read_csv('dogs.csv')
```

```python
dogs = dogs.rename(columns={'area_of_origin': 'area_of_origin_text'})
```

```python
dog_id_to_fams = {idx+1: fam.split(',') for idx, fam in enumerate(list(dogs['family']))}
```

```python
fams = ['Barbichon','Bull','Companion','Flock guard','Flockguard','Gundog','Hairless','Herding','Likestock dog','Livestock','Livestock dog','Mastiff','Mountain dog','Northern','Oriental','Pinscher','Pointer','Primative','Primitive','Retriever','Scenthound','Schnauzer','Setter','Sheepdog','Sighthound','Spaniel','Spitz','Terrier','Unknown','Versatile hunting dog','Water dog']
```

```python
id_to_fam = {fam:idx+1 for idx, fam in enumerate(fams)}
```

```python
many_fams = []
for dog_id, family_list in dog_id_to_fams.items():
    for family in family_list:
        db_data = (dog_id, id_to_fam.get(family.strip()),)
        many_fams.append(db_data)
```

```python
area_to_id = {area:idx+1 for idx, area in enumerate(areas_db)}
```

```python
dogs['area_of_origin'] = None
```

```python
dogs['area_of_origin'] = dogs.area_of_origin_text.map(lambda x: area_to_id.get(x))
```

```python
dogs['category'] = None
```

```python
categories = [('Hound'),
       ('Non-sporting'),
       ('Terrier'),
       ('Sporting'),
       ('Toy'),
       ('Herding'),
       ('Working')]
```

```python
category_to_id = {cat:idx+1 for idx, cat in enumerate(categories)}
```

```python
dogs['category'] = dogs.type.map(lambda x: category_to_id.get(x))
```

```python
dogs['weight_range'] = None
```

```python
def get_range(x):
    min_str = str(x[0])
    max_str = str(x[1])
    return "[{},{}]".format(min_str, max_str)
```

```python
dogs['weight_range'] = dogs[['weight_min', 'weight_max']].apply(get_range, axis=1)
```

```python
dogs['height_range'] = dogs[['height_min', 'height_max']].apply(get_range, axis=1)
```
